```python
from pathlib import Path
import pandas as pd
```

### Finding newly preprocessed files

```python
CSV_PATH = Path("preprocessing_progress.csv")

ANALYSIS_ROOT = Path(r"Y:\Voltage\VisualConsciousness\Analysis\Visual")
TARGET_FILE = "cG_unmixedTR_dFF.h5"


def main():
    if not CSV_PATH.exists():
        raise FileNotFoundError(f"Could not find {CSV_PATH.resolve()}")

    df = pd.read_csv(CSV_PATH)

    required = {"mouse", "day", "recording", "preprocessed"}
    if not required.issubset(df.columns):
        raise ValueError(f"CSV must contain columns: {required}")

    # Ensure boolean
    df["preprocessed"] = df["preprocessed"].astype(bool)

    mask = df["preprocessed"] == False
    to_check = df[mask]

    if to_check.empty:
        print("All recordings already marked as preprocessed.")
        return

    changed = 0

    for idx, row in to_check.iterrows():
        mouse = str(row["mouse"])
        day = str(row["day"])
        recording = str(row["recording"])

        rec_dir = ANALYSIS_ROOT / mouse / day / recording
        final_file = rec_dir / TARGET_FILE

        if final_file.is_file():
            df.at[idx, "preprocessed"] = True
            changed += 1
            print(f"UPDATED: {mouse} / {day} / {recording}")

    if changed > 0:
        df.to_csv(CSV_PATH, index=False)
        print(f"\nSaved updates to {CSV_PATH} ({changed} new recordings marked preprocessed).")
    else:
        print("No new preprocessed files found.")

    return df


if __name__ == "__main__":
    main()
```

### Finding V1Map and HCDiag Files

```python
root = Path(r"Y:\Voltage\VisualConsciousness\Raw\Visual")

# Lists of "mouse/day/recording" labels
vdt_recordings = []
hcdiag_recordings = []   # (you wrote HDdiag once; filenames use HCdiag)
v1map_recordings = []

for mouse_dir in root.iterdir():
    if not mouse_dir.is_dir():
        continue
    for day_dir in mouse_dir.iterdir():
        if not day_dir.is_dir():
            continue
        for rec_dir in day_dir.iterdir():
            if not rec_dir.is_dir():
                continue

            # only the first .dcimg file in this recording folder
            first_dcimg = next(rec_dir.glob("*.dcimg"), None)
            if first_dcimg is None:
                continue

            name = first_dcimg.name
            label = f"{mouse_dir.name}/{day_dir.name}/{rec_dir.name}"

            if "VDT" in name:
                vdt_recordings.append(label)
            elif "HCdiag" in name or "HDdiag" in name:
                hcdiag_recordings.append(label)
            elif "V1map" in name:
                v1map_recordings.append(label)

# Optional: sort for stable output
vdt_recordings.sort()
hcdiag_recordings.sort()
v1map_recordings.sort()

print(f"VDT ({len(vdt_recordings)}):")
print(vdt_recordings)
print(f"\nHCdiag/HDdiag ({len(hcdiag_recordings)}):")
print(hcdiag_recordings)
print(f"\nV1map ({len(v1map_recordings)}):")
print(v1map_recordings)
```
